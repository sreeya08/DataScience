# <font color = "green">Approach 4 : Using Imputer</font>

*Strategy to handle missing values can be :*
- mean
- median
- most_frequent

```python
from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer_median = SimpleImputer(missing_values=np.nan, strategy="median")
```

```python
imputed_data_mean = imputer_mean.fit_transform(data[cols])
imputed_data_median = imputer_median.fit_transform(data[cols])
```

```python
imputed_data_mean = pd.DataFrame(data = imputed_data_mean, columns = cols)
imputed_data_median = pd.DataFrame(data = imputed_data_median, columns = cols)
```
